## Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
print(x[:5])
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
y = sigmoid(x)

plt.plot(x,y)
plt.grid(True)
plt.show()
```

## Tanh Function

```python
def tanh(x):
    return np.tanh(x)
```

```python
y = tanh(x)

plt.plot(x,y)
plt.grid(True)
plt.show()
```

## Relu Function

```python
def relu(x):
    return np.maximum(0,x)
```

```python
y = relu(x)

plt.plot(x,y)
plt.grid(True)
plt.show()
```

## Compare All three

```python
plt.figure(figsize=(10,6))

plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, tanh(x), label="Tanh")
plt.plot(x, relu(x), label="ReLU")

plt.legend()
plt.grid(True)

plt.title("Activation Function Comparison")
plt.show()
```
