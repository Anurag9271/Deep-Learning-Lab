```python
import numpy as np

X = np.array([[1,2]])

print("Input:")
print(X)

print("Shape:",X.shape)
```

```python
W1 = np.array([
    [0.1,0.2],
    [0.3,0.4]
])

b1 = np.array([[0.1,0.1]])

print("W1 Shape:",W1.shape)
print("b1 Shape:",b1.shape)
```

```python
Z1 = np.dot(X,W1) + b1

print("Z1:")
print(Z1)  # Pre activsation output

print("Shape:",Z1.shape)
```

```python
# Activation function

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
A1 = sigmoid(Z1)

print("A1:")
print(A1)
print("Shape:",A1.shape)

# Use of activation function is that it can learn curves, shapes, images in data apart from linear patterns
```

```python
W2 = np.array([
    [0.5],
    [0.6]
])

b2 = np.array([[0.1]])
```

```python
Z2 = np.dot(A1,W2)+b2

print(Z2)
print("Shape:",Z2.shape)
```

```python
A2 = sigmoid(Z2)

print(A2)
```
